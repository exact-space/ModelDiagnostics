```python
import mlflow

MLFLOW_TRACKING_URI = "http://127.0.0.1:5000"
mlflow.set_tracking_uri(MLFLOW_TRACKING_URI)
```

```python
import contextlib

import matplotlib.pyplot as plt

@contextlib.contextmanager
def figure(*args, **kwargs):
    fig = plt.figure(*args, **kwargs)
    yield fig
    plt.close(fig)
```

```python
from sklearn.metrics import log_loss, accuracy_score, precision_score, recall_score, roc_auc_score, f1_score, fbeta_score, \
                            balanced_accuracy_score, \
                            precision_recall_fscore_support, \
                            confusion_matrix, classification_report, \
                            roc_curve, precision_recall_curve, auc, ConfusionMatrixDisplay

def get_metrics(y_test, y_pred):
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    f2 = fbeta_score(y_test, y_pred, beta=2)
    balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
    return {
        f'accuracy': round(acc, 2), 
        f'precision': round(prec, 2), 
        f'recall': round(recall, 2), 
        f'f1': round(f1, 2), 
        f'f2': round(f2, 2), 
        f'balanced_accuracy': round(balanced_accuracy, 2),
    }
```

```python
def get_or_create_experiment(project, experiment_category, experiment_for):
    experiment_name = f"{project}/{experiment_category}/{experiment_for}"
    mlflow.set_experiment(experiment_name)

def set_tags(run_tags):
    if run_tags is not None:
        mlflow.set_tags(run_tags)

def set_params(params):
    for name in params:
        mlflow.log_param(name, params[name])

def set_metrics(y_test, y_pred):
    metrics = get_metrics(y_test, y_pred)
    for name in metrics:
        mlflow.log_metric(name, metrics[name])
```
